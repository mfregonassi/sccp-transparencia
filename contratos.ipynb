```python
import sys
!{sys.executable} -m pip install requests
import requests
import json
import math
from datetime import date
```

```python
# Extracting data

club = 78 # Corinthians
gender = 'M'
professional = []

page_num = 0
while(True):
    response = requests.get(f'https://futebolpaulista.com.br/Handlers/Atleta/ListarAtletas.ashx?IdClube={club}&pageNum={page_num}')
    json_response = json.loads(response.text)
    pagination = json_response['Pagination']
    if pagination == False:
        break
    players = json_response['Retorno']
    for player in players:
        if player['Sexo'] == gender:
            player_id = player['IdAtleta']
            player_cat = player['Categoria']
            response2 = requests.get(f'https://futebolpaulista.com.br/Handlers/Atleta/ReadAtleta.ashx?IdAtleta={player_id}')
            json_response2 = json.loads(response2.text)
            player_data = json_response2['Retorno']
            age = player_data['Anos']
            year_age = int(player_data['DataNascimento'][-4:])
            if player_cat == 'P' or age >= 16:
                if year_age >= 2004:
                    player_data['Grupo'] = 'S17'
                elif year_age >= 2001:
                    player_data['Grupo'] = 'S20'
                elif year_age >= 1998:
                    player_data['Grupo'] = 'S23'
                else:
                    player_data['Grupo'] = '23+'
                player_data['Categoria'] = player_cat
                professional.append(player_data)
    page_num += 1
    
with open(f'./data/{date.today().strftime("%Y%m%d")}.json', 'w', encoding='utf-8') as f:
    json.dump(professional, f, ensure_ascii=False, indent=4)
```

```python
with open(f'./data/{date.today().strftime("%Y%m%d")}.json') as f:
    players = json.load(f)

u_17_p = 0
u_17_a = 0
u_20_p = 0
u_20_a = 0
u_23_p = 0
u_23_a = 0
o_23_p = 0
o_23_a = 0

for player in players:
    cat = player['Categoria']
    group = player['Grupo']
    if cat == 'A':
        if group == 'S17':
            u_17_a += 1
        elif group == 'S20':
            u_20_a += 1
        elif group == 'S23':
            u_23_a += 1
        elif group == '23+':
            o_23_a += 1
    elif cat == 'P':
        if group == 'S17':
            u_17_p += 1
        elif group == 'S20':
            u_20_p += 1
        elif group == 'S23':
            u_23_p += 1
        elif group == '23+':
            o_23_p += 1
                 
u_17 = u_17_p + u_17_a
u_20 = u_20_p + u_20_a
u_23 = u_23_p + u_23_a
o_23 = o_23_p + o_23_a
num_a = u_17_a + u_20_a + u_23_a + o_23_a
num_p = u_17_p + u_20_p + u_23_p + o_23_p
num_players = num_a + num_p
```

```python
next_6 = ['03/2021', '04/2021', '05/2021', '06/2021', '07/2021', '08/2021']
expiring_0 = []
expiring_1 = []
expiring_2 = []
expiring_3 = []
expiring_4 = []
expiring_5 = []
expiring_21 = []
expiring_22 = []
expiring_23 = []
expiring_24 = []
expiring_25 = []
expiring_26 = []

for player in players:
    contract_end = player['DataTerminoContrato']
    if contract_end.endswith(next_6[0]):
        expiring_0.append(player)
    elif contract_end.endswith(next_6[1]):
        expiring_1.append(player)
    elif contract_end.endswith(next_6[2]):
        expiring_2.append(player)
    elif contract_end.endswith(next_6[3]):
        expiring_3.append(player)
    elif contract_end.endswith(next_6[4]):
        expiring_4.append(player)
    elif contract_end.endswith(next_6[5]):
        expiring_5.append(player)
    elif contract_end.endswith('2021'):
        expiring_21.append(player)
    elif contract_end.endswith('2022'):
        expiring_22.append(player)
    elif contract_end.endswith('2023'):
        expiring_23.append(player)
    elif contract_end.endswith('2024'):
        expiring_24.append(player)
    elif contract_end.endswith('2025'):
        expiring_25.append(player)
    elif contract_end.endswith('2026'):
        expiring_26.append(player)
    else:
        continue
```

```python
print(f'{num_players} jogadores acima de 16 anos sob contrato')
print(f'{num_p} com contratos profissionais, {num_a} amadores')
print('')
print(f'{u_17} sub-17, {u_17_p} com contratos profissionais, {u_17_a} amadores')
print(f'{u_20} sub-20, {u_20_p} com contratos profissionais, {u_20_a} amadores')
print(f'{u_23} sub-23, {u_23_p} com contratos profissionais, {u_23_a} amadores')
print(f'{o_23} acima de 23, {o_23_p} com contratos profissionais, {o_23_a} amadores')
print('')
print(f'Expirando em {next_6[0]}:')
for player in expiring_0:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')    
print(f'Expirando em {next_6[1]}:')
for player in expiring_1:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')    
print(f'Expirando em {next_6[2]}:')
for player in expiring_2:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')
print(f'Expirando em {next_6[3]}:')
for player in expiring_3:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')
print(f'Expirando em {next_6[4]}:')
for player in expiring_4:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')
print(f'Expirando em {next_6[5]}:')
for player in expiring_5:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')
print('---')
print(f'Outros {len(expiring_21)} contratos expirando em 2021')
print(f'{len(expiring_22)} contratos expirando em 2022')
print(f'{len(expiring_23)} contratos expirando em 2023')
print('')
print(f'{len(expiring_24)} contratos expirando em 2024:')
for player in expiring_24:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')
print(f'{len(expiring_25)} contratos expirando em 2025:')
for player in expiring_25:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
print('')
print(f'{len(expiring_26)} contratos expirando em 2026:')
for player in expiring_26:
    name = player['Nome']
    age = player['Anos']
    contract_end = player['DataTerminoContrato']
    print(f'{name} ({age}) - {contract_end}')
```
